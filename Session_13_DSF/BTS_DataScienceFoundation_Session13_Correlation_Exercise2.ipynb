### Exercise 2 -  Calculating Correlation in Pandas


We will be using a dataset "Boston house prices". Calculate the Pearson correlation between "MEDV-Median value of owner-occupied homes in $1000's)" and all the variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets 
data = datasets.load_boston()
import statsmodels.api as sm
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
target = pd.DataFrame(data.target, columns=["MEDV"]) 
target.head()
```

```python
print (data.DESCR)
```

```python
df2 = df.assign(target=target.values)
df2.head()
```

```python
df2.rename(columns={'target': 'MEDV'}, inplace=True)
df2.head()
```

```python
df2.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

## RM, LSTAT, PTRATIO are the ones that correlate with a value above .5 so they are good x candidates. They don't have that high a correlation so I would use all 3 in my regresion model

```python
X1 = df2[['RM','LSTAT','PTRATIO']]
y1 = df2['MEDV']
X1 = sm.add_constant(X1) 
model1 = sm.OLS(y1, X1).fit()
model1.summary()
```
