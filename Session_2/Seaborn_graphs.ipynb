```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/roman2006/PycharmProjects/Liana Clone/Session_2/2_titanic_dataset.csv")
```

```python
plt.figure(figsize=[12,10])
sns.countplot(df['Survived'])
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.barplot('Sex','Survived',data=df)
plt.show()
```

```python
sns.pointplot(x='Parch', y='Survived', hue='Sex', data=df, palette={"male":'blue','female':'red'})
plt.show()
```

```python
plt.figure(figsize=[12,12])
sns.swarmplot(x='Sex',y='Age',data=df)
plt.show()
```

```python
plt.figure(figsize=[12,10])
sns.factorplot('Pclass','Survived','Sex', data= df, kind='bar',palette='muted', legend=False)
plt.show()
```

```python
g=sns.FacetGrid(df, col='Sex', row='Pclass',margin_titles=True)
g.map(plt.scatter,'Fare','Age')
# add title to the figure
g.fig.suptitle('Plot')
plt.show()
```
