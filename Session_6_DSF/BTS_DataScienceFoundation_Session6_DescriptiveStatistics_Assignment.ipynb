## Finding outliers

### Age

```python
df['Age_std']=(((df['Age']-df['Age'].mean()).apply(abs)/df['Age'].std()))
df.sort_values(by='Age_std',ascending=False).head(100)
```

### Since we modified the missing 177 values with the mean, most values became outliers, since now there is so much data concentrated on the mean.

### Fare

```python
df['Fare_std']=(((df['Fare']-df['Fare'].mean()).apply(abs)/df['Fare'].std()))
df.sort_values(by='Fare_std',ascending=False).head(50)
```

### We can see that since most people payed little money to get on the Titanic most people who payed more than 100 are outliers
