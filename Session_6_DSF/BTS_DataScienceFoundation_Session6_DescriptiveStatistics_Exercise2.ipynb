## Question: What are the five regions with the lowest fertility rate and which have a outlier behavior?

```python
df['DatoN_std']=(((df['DatoN']-df['DatoN'].mean()).apply(abs)/df['DatoN'].std()))
df.sort_values(by='DatoN',ascending=True).head(5)
```

#### Those are the places with the lowest fertility rates

```python
df.sort_values(by='DatoN_std',ascending=False).head(8)
```

#### This are the 8 possible outliers since they all have a value above 1.5 in their standard deviation
